```python
#! pip install spotipy --upgrade
import json
import pandas as pd
from spotify_utils import get_playlist_df
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from etl_utils import get_unique_map_from_df_col, clean_date
```

```python
file = open('credentials.json')
credentials = json.load(file)
file.close()

#credentials
```

```python
auth_manager = SpotifyClientCredentials(client_id = credentials["client_id"], 
                                        client_secret = credentials["client_secret"])

sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
file = open('playlist_info.json')
uri_map = json.load(file)
file.close()

uri_map
```

```python
%%time

df_map = {}
for artist_playlist, uri in uri_map.items():
    
    print("I'm on artist "+ str(artist_playlist))
    
    if artist_playlist in df_map:
        continue
    df = get_playlist_df('spotify', uri, sp, song_limit=30)
    
    df_map[artist_playlist] = df
```

```python
df_list = []
for key, value in df_map.items():
    
    df = df_map[key]
    df['artist_playlist'] = key
    df_list.append(df)
    
df_full = pd.concat(df_list, ignore_index=True)
```

```python
df_full.to_excel("spotify_data.xlsx", index=False)
```
